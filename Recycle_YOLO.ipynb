# 4. train-test split

```
from glob import glob
img_list = glob('/content/gdrive/My Drive/Recycle_image/*.jpg')
```

```
len(img_list)
```

```
from sklearn.model_selection import train_test_split

train_img_list, test_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)
```

```
with open('/content/gdrive/My Drive/Recycle_image/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/gdrive/My Drive/Recycle_image/test.txt', 'w') as f:
  f.write('\n'.join(test_img_list) + '\n')  
```
